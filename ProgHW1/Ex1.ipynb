```python
import numpy as np

matrix_0 = np.array([[2,4,6], [7,3,5]])
print(matrix_0)

assert np.shape(matrix_0) == (2,3)
```

```python
print(matrix_0[0, :])
print(matrix_0[:, 1])
```

```python
new_row = np.zeros((1, matrix_0.shape[1]))
matrix_1 = np.vstack((matrix_0, new_row))
print(matrix_1)
```

```python
import matplotlib.pyplot as plt

xvalues = np.linspace(0, 3, num=100)
yvalues = [x**2 for x in xvalues]
plt.plot(xvalues, yvalues)
plt.show()
```
